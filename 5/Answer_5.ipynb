```python
import pandas as pd
data = pd.read_csv('abalone.csv')
data.head()
```

```python
data.describe()
```

```python
from scipy import stats
```

```python
stat, p = stats.shapiro(data['Rings']) # тест Шапиро-Уилк
print('Statistics=%.3f, p-value=%.3f' % (stat, p))
alpha = 0.05
if p > alpha:
    print('Принять гипотезу о нормальности')
else:
    print('Отклонить гипотезу о нормальности')
```

В качестве критического значения в большинстве случаев берется 0.05. При p-значении меньше 0.05 мы вынуждены отклонить нулевую гипотезу. В результате мы получили низкое p-значение и, следовательно, отклоняем нулевую гипотезу.
