# Вывод
Мы выяснили, что показатели лет в компании при условии, что StockOptionLevel = 1 или = 0, не сильно отличаются, кроме показателя от 0 до 1 года.

```python
df_hr['anomally_sal'] = np.where(
                            df_hr['YearsAtCompany'] > 
                            np.mean(df_hr.YearsAtCompany) + 3*np.std(df_hr.YearsAtCompany),                 
                            1, 0)
```
